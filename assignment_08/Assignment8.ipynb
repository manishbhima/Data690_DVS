- Cleansing the data to make the year, population, cost, cost per person numerical type
- Code functions to cleanse the columns. One function for one column
- Using apply() method of data frame to apply the functions to the columns for data cleansing

```python
# Function to cleanse the year

def cleanse(a_string):
    return int(a_string.replace("*",""))



df["YEAR"] = df["Census Year"].apply(cleanse)

df
```

```python
# Function to convert a str of numbers separated by commas into a integer

def cleanse1(a_string):
    return int(a_string.replace(",",""))


df["POPULATION"] = df["Total Population"].apply(cleanse1)

df

df.info()
```

```python
# Function to convert the cost which is seperated by commas to a numeric type
def converter(column_value):
  #return column_value.replace('$', '')
  #return column_value.replace(',', '')
  return column_value.replace(' Billion','000000000')
  

def converter1(column_value):
  return column_value.replace('$', '')

def converter2(column_value):
  return column_value.replace(',', '')



df["Census Cost"] = df["Census Cost"].apply(converter)
df

df["Census Cost"] = df["Census Cost"].apply(converter1)
df

df["COST"] = df["Census Cost"].apply(converter2)
df


```

```python
# Converting the data type from object to float
df[df.columns[6]] = df[df.columns[6]].astype(float)
df
```

```python
df.info()
```

```python
# Function to convert the Average cost which is seperated by commas to numeric type
import re
def cleanse_currency(x):
  for i in x:
    z= re.search('^\$', i)
    if(z!=None):
      cent = (int(round(float(x.strip('$'))*100)))
      return (cent)
    else: 
      return (x.replace(' cents',''))
  


df["COST PER PERSON"] = df["Average Cost Per Person"].apply(cleanse_currency)

df
```

```python
# Converting the datatype from object to float
df[df.columns[7]] = df[df.columns[7]].astype(float)
df
```

```python
df.info()
```

```python
# Adding the cleansed data to a new dataframe
df1 = df[["YEAR",	"POPULATION",	"COST",	"COST PER PERSON"]]
df1
```

- Using Plotly, plotting the average cost per person over time.

```python
import plotly.express as px  
```

```python
fig = px.bar(
    df1, 
    x = "YEAR", 
    y = "COST PER PERSON", 
    title = "Average Cost Per Person over Time", 
    color = "YEAR", 
    template = "plotly_dark"
)

fig.update_layout(showlegend=False)

fig.show()
```

- Using Plotly, plotting the scatter plot of Total Population vs Total Cost.

```python
fig = px.scatter(
    df1,
    x="POPULATION",
    y="COST", 
    title="Total Population vs Total Cost", 
    color="YEAR", 
    template="presentation"
)

fig.update_layout(showlegend=True)
fig.update_layout(
    xaxis_title="Total Population",
    yaxis_title="Total Cost",)


fig.show()
```
